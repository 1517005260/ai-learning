# 文本嵌入

![embed](./img/embed.png)

如上图所示，文本嵌入就是把一句话通过模型转换成向量，这里需要注意的是，每个模型最后生成的向量维度可能是不一样的。文本嵌入的目的，就是做相关的文本检索，根据句向量的相似度，找到相关的文本。

```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI()
```

```python
# input是单条数据的情形

response = client.embeddings.create(
    model="m3e",    # 根据one-api定
    input=["感冒了吃什么药好得快？"]   # m3e 接收列表
)

response
```

```python
len(response.data[0].embedding)   # 向量维度
```

```python
# input是多条数据的情形

response = client.embeddings.create(
    model="text-embedding-ada-002",
    input=["感冒了吃什么药好得快？", "阿莫西林能治感冒吗？"]
)

for item in response.data:
    print(len(item.embedding))
```

由于转化之后向量的维度，不同的模型输出的维度是不同的，所以在一个项目中，只会使用一种模型进行编码，才能保证维度的一致性，否则维度不一致可能会报错。
