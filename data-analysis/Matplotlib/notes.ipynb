# 1 PyPlot

大多数的 Matplotlib 的实用函数都在 pyplot 下

```python
import matplotlib.pyplot as plt
```

## 1.1 简单直线

```python
# 画一条 (0,0) -> (6,250) 的直线
import numpy as np
x = np.array([0,6]) # x1,x2
y = np.array([0,250]) # y1,y2

plt.plot(x,y)
plt.show()
```

## 1.2 无线图

```python
plt.plot(x,y,'o')
plt.show()
```

## 1.3 折线图

```python
x = np.array([1,2,3,4]) 
y = np.array([5,1,2,8])

plt.plot(x,y)
plt.show()
```

注意：不传入 x 的坐标，则默认 x 为[0,1,2,3,4,...]

# 2 标记

## 2.1 标记 点、线、颜色

```python
y = np.array([1,2,3,4])
plt.plot(y, marker='o')
plt.show()
```

```python
plt.plot(y, marker='*')
plt.show()
```

```python
plt.plot(y, 'o:r') # 标记点、虚线、红色
plt.show()
```

## 2.2 标记尺寸

```python
plt.plot(y, marker='*',ms=20)
plt.show()
```

此外，还有很多线标记、颜色标记等，和css一样，现用现查即可。

# 3 线

## 3.1 线型

```python
plt.plot(y,linestyle='dotted')
plt.show()
```

```python
plt.plot(y,ls='--') # 简写
plt.show()
```

## 3.2 多行

```python
x1 = np.array([1,3,5,7])
y1 = np.array([3,8,1,10])
x2 = np.array([2,4,6,8])
y2 = np.array([6,2,7,11])

plt.plot(x1,y1,x2,y2)

plt.show()
```
